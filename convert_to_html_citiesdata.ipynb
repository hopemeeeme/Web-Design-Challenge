```python
# Dependencies
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# Read in CSV.
df = pd.read_csv('cities.csv')
df.head(10)
```

```python
df.dtypes
```

```python
# Convert INT64 Date to Datetime
df['Date'] = pd.to_datetime(df['Date'], unit='s')
df.head(10)
```

```python
# Output as HTML, drop index.
df.to_html('data_all.html', index=False)
```

```python
# Create scatter plot for Lat vs Max Temp.
plt.scatter(df['Lat'], df['Max Temp'], marker='o', s=20, c='b', edgecolors='green', alpha=.75)

# Format axis/labels.
plt.title('City Latitude vs. Max Temperature - 04/13/2020')
plt.ylabel('Max Temperature (F)')
plt.xlabel('Latitude')
plt.grid(True)
plt.tight_layout()

# Save the figure.
plt.savefig('assets/images/Fig1.png')

plt.show()
```
